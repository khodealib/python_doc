# Use Python Interpreter
## Installing Python

You can use the following links to install Python (it is better to use version 3.9).

- [Linux](https://linuxize.com/post/how-to-install-python-3-9-on-ubuntu-20-04/)
- [Windows](https://python.tutorials24x7.com/blog/how-to-install-python-3-9-on-windows-10)
- [MacOS](https://python.tutorials24x7.com/blog/how-to-install-python-3-9-on-mac)

After installing Python, the following Python interpreter is executed by writing the following command in the context of each operating system (terminal on Linux and Mac, command line in Windows).
```bash
$ python3
```
